## Convert to grayscale

### Subtask:
Convert to grayscale


**Reasoning**:
Download the image from a public URL, load it, convert it to grayscale, and display the grayscale image.



```python
import requests

image_url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.png"
image_filename = "lena.png"

try:
    response = requests.get(image_url, stream=True)
    response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)

    with open(image_filename, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)

    print(f"Image downloaded successfully as {image_filename}")

    img = cv2.imread(image_filename)

    if img is not None:
        imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        cv2_imshow(imgGray)
    else:
        print("Error: Image downloaded but could not be read by OpenCV.")

except requests.exceptions.RequestException as e:
    print(f"Error downloading image: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

**Reasoning**:

## Apply blur

### Subtask:

**Reasoning**:
